### Imports

```python
import cv2

from matplotlib import pyplot as plt
from ultralytics import YOLO

import os    
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
```

### Edge Detection

```python
chessboard_img = cv2.cvtColor(cv2.imread("chessboard.jpg"), cv2.COLOR_BGR2RGB)
plt.imshow(chessboard_img)
```

```python
model = YOLO('yolov8n.pt')
results = model.train(data="data.yaml", imgsz=640, epochs=100, batch=8, name="corner-detector")
```

```python
model = YOLO("runs/detect/corner-detector/weights/best.pt")
results = model.predict(source='chessboard.jpg', line_width=3, conf=0.25, save_txt=True, save=True)
```
